```python
with open('store_data.csv', 'r')as file:
    lines = file.readlines()
```

```python
records = []
for i in lines:
    a =i.strip()
    cart = a.strip()
    cart = a.split(',')
    records.append(cart)
```

```python
records[0:5]
```

```python
len(records)
```

Data extraction completed

```python
sup =(7*5)/len(records)
sup
```

```python
conf =0.2
```

```python
lift =3
```

```python
from apyori import apriori
association_rules = apriori(records,min_support =sup,min_confidence=conf,min_left=3)
association_results = list(association_rules)
```

```python
association_results
```

```python
association_results[0][0]
```

```python
freq_bought =[]
for i in association_results:
    freq_bought.append(i[0])
```

Saving the association results in text file

```python
with open ('Results.txt','w')as file2:
    for i in  association_results:
        file2.write(str(i)+'\n')
```
